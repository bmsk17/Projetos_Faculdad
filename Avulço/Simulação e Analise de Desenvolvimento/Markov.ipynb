```python
import random
import numpy as np

# Definição dos estados
states = ["Ensolarado", "Nublado", "Chuvoso"]
num_states = len(states)

# Matriz de Transição (exemplo)
transition_matrix = np.array([
    [0.7, 0.2, 0.1],
    [0.3, 0.4, 0.3],
    [0.1, 0.3, 0.6]
])

# Escolha Inicial do Estado
initial_state = np.random.choice(states, p=[0.5, 0.3, 0.2])

# Simulação
num_steps = 30
current_state = initial_state
simulated_states = [current_state]

for _ in range(num_steps):
    next_state = np.random.choice(states, p=transition_matrix[states.index(current_state)])
    simulated_states.append(next_state)
    current_state = next_state

# Resultados da Simulação
print("Simulação da Cadeia de Markov:")

i = 0
for elemento in simulated_states:
    i=i+1
    print(i," --> ",elemento)
```

```python
import numpy as np

# Definição dos estados
states = ["Vitoria", "Empate", "Derrota"]
num_states = len(states)

# Matriz de Transição corrigida
transition_matrix = np.array([
    [0.529, 0.118, 0.353],
    [0.643, 0.214, 0.143],
    [0.357, 0.071, 0.572]
])

# Realizar 10 simulações
num_simulations = 10
num_steps = 20

for simulation in range(num_simulations):
    print(f"Simulação {simulation + 1}:")
    
    # Escolher um estado inicial aleatório
    initial_state = np.random.choice(states)
    current_state = initial_state
    simulated_states = [current_state]
    
    # Contadores de vitórias, empates e derrotas
    vitorias = 0
    empates = 0
    derrotas = 0
    
    for _ in range(num_steps):
        next_state = np.random.choice(states, p=transition_matrix[states.index(current_state)])
        simulated_states.append(next_state)
        current_state = next_state
        
        # Atualizar contadores
        if next_state == "Vitoria":
            vitorias += 1
        elif next_state == "Empate":
            empates += 1
        elif next_state == "Derrota":
            derrotas += 1
    
    total = vitorias + empates + derrotas
    pct_vitorias = (vitorias / total) * 100
    pct_empates = (empates / total) * 100
    pct_derrotas = (derrotas / total) * 100
    
    print(f"Vitórias: {pct_vitorias:.2f}% | Empates: {pct_empates:.2f}% | Derrotas: {pct_derrotas:.2f}%")
    print()
```
